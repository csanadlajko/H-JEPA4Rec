# <i>Hierarchical Joint Embedding Predictive Architecture for Session Based Recommendation</i>

## Differentiate two types:
- H-JEPA without uncertainty
- H-JEPA with uncertainty handling

### 1.<b> H-JEPA without uncertainty handling</b>
<img src="h_jepa.png" width="1000" height="500" />

(credit: https://openreview.net/pdf?id=BZ5a1r-kVsf)

- An architecture used for acquiring different level of abstraction from the same situations
- Eg. when browsing an item, we associate it with the following levels:
    - Level 0: the item's representation itself
    - Level 1: the brand and it's different items
    - Level 2: the occasion that the item's bought for
    - Level 3: contextual - the reasing why the user is browsing the item
    - ...
- Does not handle uncertainty in case of future representation predictions

### 2.<b> H-JEPA with uncertainty handling</b>
<img src="h_jepa_uncertain.png" width="1000" height="500" />


(credit: https://openreview.net/pdf?id=BZ5a1r-kVsf)

- Same architecure as before, with an additional uncertainty-handling mechanism
- We can create suchs abstraction levels as mentioned before
- Realistic environment are not entirely predictable
    - We add additional latent variables which hold the values of the information that cannot be derived from prior observations
- When predicting item / context representation, no more than 1 prediction block is needed
